# Feature Engineering

**Absolute Magnitude**

The dataset contains values for the Vmag which refers to the visual magnitude of a star. These are values assigned to a star based on how visible it is as viewed by a telescope, for this reason it is also sometimes called as apparent magnitude.

To gain a better understanding of the properties of a star we are going to convert this data to absolute magnitude. The absolute magnitude is the magnitude of a star as measured from a standard distance of 10 parsecs.

The formula to convert visual magnitude (vmag) to absolute magnitude (abmag) is as follows:

\begin{align}
  M &= m - 5  log_{10} (\frac{d}{10})
\end{align}

But in order for us to use this formula, we need to first determine the distance of the star in parsec. This distance can be obtained from the parallax values using the formula provided below:

\begin{align}
  d=1/ω
\end{align}

```python
#Create a new column with the distances obtained from the parallax column
#Note that the parallax recored were in milliarcseconds so to get the proper
#unit for the distance, multiply it by 1000


def distance(df):
  distance = 1000/df['Plx']
  df.insert(2, 'Distance', distance)
  return df
```

```python
#Create a new column for the absolute value using the distance and vmag

def abmag(df):
  abmag = df['Vmag'] - 5*np.log10(df['Distance']/10)
  df.insert(5, 'Abmag', abmag)
  return df
```

**Effective Temperature**

The dataset contains B-V color values which is a way of quantifying the color of stars using two different filters. One filter is blue (B) filter that only lets a narrow range of colors or wavelengths through centered on the blue colors, and the other is a visual (V) filter that only lets the wavelengths close to the green-yellow band through.

The B-V value as have already been discussed above is related to the stars temperature and thus this is something that we can use to determine the effective temperature. This can be done using what is known as the Ballestero's Formula as shown below:

\begin{align}
  T = 4600K \left(\frac{1}{0.92(B-V) + 1.7} + \frac{1}{1.5(B-V) + 0.62} \right)
\end{align}

```python
#Create a new column for the effective temeprature of the stars

def efftemp(df):
  efftemp = 4600*(1/(0.92*df['B-V']+1.7) + 1/(1.5*df['B-V'] + 0.62))
  df.insert(5, 'EffTemp', efftemp)
  return df
```

**Transformation Pipeline**

```python
#Using the functions defined above, make Function Transformers using sklearn

from sklearn.preprocessing import FunctionTransformer

dist_trans = FunctionTransformer(distance)
abmag_trans = FunctionTransformer(abmag)
efftemp_trans = FunctionTransformer(efftemp)
```

```python
#Construct the transformation pipeline

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

x_train = train_set.drop('SpClass', axis=1)
y_train = train_set['SpClass'].copy()

#seperate the numerical attributes from the categorical attributes
num_attr = x_train.select_dtypes('float', 'int').columns
cat_attr = x_train.select_dtypes('object').columns

custom_trans = Pipeline([
    ('dist', dist_trans),
    ('abmag', abmag_trans),
    ('efftemp', efftemp_trans)
])


num_pipe = Pipeline([
    ('dist', dist_trans),
    ('abmag', abmag_trans),
    ('efftemp', efftemp_trans),
    ('Standard Scalar', StandardScaler())
])

OH_pipe = Pipeline([
    ('oh', OneHotEncoder(handle_unknown='ignore'), cat_attr)
])

full_pipe = ColumnTransformer([
    ('num', num_pipe, num_attr),
    ('cat', OneHotEncoder(), cat_attr)
])

x_train_prep = full_pipe.fit_transform(x_train)
```

```python
def label_trans(y, df):
  #this function transforms the categorical label values into ordinal data

  from sklearn.preprocessing import OrdinalEncoder

  y_df = pd.DataFrame(y, index = df.index)

  oe = OrdinalEncoder(categories=[['O', 'B', 'A','F','G','K','M']])
  y_prep = oe.fit_transform(y_df)

  return y_prep

```

```python
y_train_prep = label_trans(y_train, train_set)
```

# Model Selection

```python
#import the models to be used

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_predict
```

```python
#load the models

rf_model = RandomForestClassifier(random_state=42)
log_model = LogisticRegression(max_iter=1000)
nb_model = GaussianNB()
sgd_model = SGDClassifier()
knn_model = KNeighborsClassifier()
```

```python
#create a dict that contains the models and model names

model_dict = {
    'Random Forest Classifier': rf_model,
    'Logistic Regression': log_model,
    'Naive Bayes': nb_model,
    'SGD Classifier': sgd_model,
    'K Nearest Neighbors': knn_model
}
```

```python
#train the models given the train set and print the classification report after
#making predictions with the test set

from sklearn.metrics import classification_report

for model in model_dict:
  model_pred = cross_val_predict(model_dict[model], x_train_prep, y_train_prep.ravel(), cv=5)
  print(model, 'Classification Report \n')
  print(classification_report(model_pred, y_train_prep))
```

**Neural Networks Using Functionals**

```python
#build the neural network

input = keras.layers.Input(x_train_prep.shape[1:])
hidden1 = keras.layers.Dense(500, activation='relu')(input)
hidden2 = keras.layers.Dense(30, activation = 'relu')(hidden1)
concat = keras.layers.Concatenate()([input, hidden2])
output = keras.layers.Dense(7, activation= 'softmax')(concat)
model = keras.models.Model(inputs=[input], outputs=[output])
```

```python
model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'sgd',
    metrics = ['accuracy']
)
```

```python
history = model.fit(x_train_prep, y_train_prep, epochs = 30, validation_split = 0.2)
```
